````python
#setup env
import time
import warnings
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import cluster, datasets, mixture
from sklearn.neighbors import kneighbors_graph
from sklearn.preprocessing import StandardScaler
````

````python
#helper calls
n_samples = 1500
random_state = 170
X, y = datasets.make_blobs(n_samples=n_samples, random_state=random_state)
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X_aniso = np.dot(X, transformation)

#generating Data Sets A-F
A = datasets.make_circles(n_samples=n_samples, factor=.5,noise=.05)[0]
B = datasets.make_moons(n_samples=n_samples, noise=.05)[0]
C = datasets.make_blobs(n_samples=n_samples, random_state=8)[0]
D = np.random.rand(n_samples, 2)
E = (X_aniso, y)[0]
F = datasets.make_blobs(n_samples=n_samples,
                             cluster_std=[1.0, 2.5, 0.5],
                             random_state=random_state)[0]
````

## Exercise 1
Plot all raw data sets A-F in one figure.

Hints: use MATPLOTLIB ```scatter``` plots and ```subfigures```.

````python
fig, ax= plt.subplots(2, 3, figsize=(9,6))

ax[0,0].scatter(A[:,0], A[:,1])
ax[0,1].scatter(B[:,0], B[:,1])
ax[0,2].scatter(C[:,0], C[:,1])
ax[1,0].scatter(D[:,0], D[:,1])
ax[1,1].scatter(E[:,0], E[:,1])
ax[1,2].scatter(F[:,0], F[:,1])
````

## Exercise 2
Perform ```K-Means``` clustering on all data sets: https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html#sklearn.cluster.KMeans
* 2.1 plot all results
* 2.2 manually try to find the best $k$ for each data set (by visual evaluation)
* 2.3 Use the ```K-Means++``` initialization 

````python
from sklearn.cluster import KMeans
````

````python
A_KMeans = KMeans (n_clusters = 2, random_state=0)
B_KMeans = KMeans (n_clusters = 2, random_state=0)
C_KMeans = KMeans (n_clusters = 3, random_state=0)
D_KMeans = KMeans (n_clusters = 10, random_state=0)
E_KMeans = KMeans (n_clusters = 3, random_state=0)
F_KMeans = KMeans (n_clusters = 3, random_state=0)
````

````python
A_pred = A_KMeans.fit_predict(A)
B_pred = B_KMeans.fit_predict(B)
C_pred = C_KMeans.fit_predict(C)
D_pred = D_KMeans.fit_predict(D)
E_pred = E_KMeans.fit_predict(E)
F_pred = F_KMeans.fit_predict(F)
````

````python
fig, ax= plt.subplots(2, 3, figsize=(9,6))

ax[0,0].scatter(A[:,0], A[:,1], c= A_pred)
ax[0,1].scatter(B[:,0], B[:,1],c= B_pred)
ax[0,2].scatter(C[:,0], C[:,1],c= C_pred)
ax[1,0].scatter(D[:,0], D[:,1],c= D_pred)
ax[1,1].scatter(E[:,0], E[:,1],c= E_pred)
ax[1,2].scatter(F[:,0], F[:,1],c= F_pred)
````

## Exercise 3
Performs ```DBSCAN``` clustering on all data sets and plot all results: https://scikit-learn.org/stable/modules/generated/sklearn.cluster.DBSCAN.html#sklearn.cluster.DBSCAN


````python
from sklearn.cluster import DBSCAN
````

````python
A_db = DBSCAN(min_samples=2, eps=0.2).fit_predict(A)
B_db = DBSCAN(min_samples=2, eps=0.2).fit_predict(B)
C_db = DBSCAN(min_samples=5, eps=0.9).fit_predict(C)
D_db = DBSCAN(min_samples=20, eps=0.3).fit_predict(D)
E_db = DBSCAN(min_samples=4, eps=0.305).fit_predict(E)
F_db = DBSCAN(min_samples=10, eps=0.8).fit_predict(F)

fig, ax= plt.subplots(2, 3, figsize=(9,6))

ax[0,0].scatter(A[:,0], A[:,1], c= A_db)
ax[0,1].scatter(B[:,0], B[:,1],c= B_db)
ax[0,2].scatter(C[:,0], C[:,1],c= C_db)
ax[1,0].scatter(D[:,0], D[:,1],c= D_db)
ax[1,1].scatter(E[:,0], E[:,1],c= E_db)
ax[1,2].scatter(F[:,0], F[:,1],c= F_db)
````
