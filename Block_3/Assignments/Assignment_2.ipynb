### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
drinks_spirit=drinks['spirit_servings'].agg([np.mean,np.min,np.max])
drinks_spirit = pd.DataFrame(drinks_spirit)
drinks_spirit
```

```python
mean_spirit = drinks['spirit_servings'].mean()
max_spirit = drinks['spirit_servings'].max()
min_spirit = drinks['spirit_servings'].min()
```

```python
#create a Dataframe from a dictonary
my_dict={
            'who':['world'],
            'mean':[mean_spirit],
            'max':[max_spirit],
            'min':[min_spirit],
}
spirit_consump=pd.DataFrame(my_dict)
spirit_consump
```

```python
print(spirit_consump.index)
spirit_consump.set_index('who',inplace=True)
spirit_consump
```
